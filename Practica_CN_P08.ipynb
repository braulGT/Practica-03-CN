#Introducción
Los principios de lógica difusa se pueden utilizar para agrupar datos multidimensionales, asignando a cada punto una membresía en cada centro de grupo de 0 a 100 por ciento. Esto puede ser muy poderoso en comparación con la agrupación tradicional de umbral estricto, donde a cada punto se le asigna una etiqueta precisa y nítida. Este algoritmo funciona asignando membresía a cada punto de datos correspondiente a cada centro de grupo sobre la base de la distancia entre el centro de grupo y el punto de datos. Cuanto más cerca están los datos del centro del grupo, más es su membresía hacia el centro del grupo en particular. Claramente, la suma de miembros de cada punto de datos debe ser igual a uno.\
La agrupación de datos es el proceso de dividir los elementos de los datos en clases o clusters, de manera que los elementos de una misma clase sean lo más parecidos posible y los elementos de clases diferentes sean lo más disímiles. Dependiendo de la naturaleza de los datos y del propósito para el que se utiliza la agrupación, se pueden utilizar diferentes medidas de similitud para colocar los elementos en clases, donde la medida de similitud controla cómo se forman los clusters. \
Fuzzy c-means (FCM) es una técnica de clustering de datos en la que cada punto de datos pertenece a un cluster en cierto grado especificado por un grado de pertenencia. Este método fue desarrollado por Dunn en 1973 y mejorado por Bezdek en 1981 como una mejora de los métodos de clustering anteriores. Proporciona un método que muestra cómo agrupar puntos de datos que pueblan algún espacio multidimensional en un número específico de clusters diferentes. Es un algoritmo de agrupación en clúster no supervisado que nos permite construir una partición difusa a partir de datos. Se le considera como uno de los algoritmos de agrupación de clústeres suaves más utilizados. \
soft clustering.JPG  \
En la función fcm de la línea de comandos comienza con una estimación inicial de los centros de los clusters, que pretenden marcar la ubicación media de cada cluster. La conjetura inicial para estos centros de cluster es muy probablemente incorrecta. Además, fcm asigna a cada punto de datos un grado de pertenencia a cada clúster. Al actualizar de forma iterativa los centros de los clusters y los grados de pertenencia de cada punto de datos, fcm mueve de forma iterativa los centros de los clusters a la ubicación correcta dentro de un conjunto de datos. Esta iteración se basa en la minimización de una función objetivo que representa la distancia de cualquier punto de datos a un centro de cluster ponderado por el grado de pertenencia de ese punto de datos.\
Puede utilizar la información devuelta por fcm para ayudarle a construir un sistema de inferencia difusa creando funciones de pertenencia para representar las cualidades difusas de cada cluster. Para generar un sistema de inferencia difusa de tipo Sugeno que modele el comportamiento de los datos de entrada/salida, puede configurar el comando genfis para que utilice el clustering FCM. \
En el agrupamiento Fuzzy-C Means, cada punto tiene una ponderación asociada con un grupo en particular, por lo que un punto no se ubica "en un grupo" tanto como tiene una asociación débil o fuerte con el grupo, que está determinada por la distancia inversa. al centro del grupo. La agrupación difusa de c-means se lleva a cabo a través de skfuzzy.cmeans, y la salida de esta función puede ser reutilizada para clasificar nuevos datos de acuerdo con los clusters calculados (también conocido como predicción) a través de skfuzzy.cmeans_predict\
La función fuzzy c-means es un caso especial de K-means cuando la función de probabilidad utilizada es simplemente 1 si el punto de datos está más cerca de un centroide y 0 en caso contrario.\

El flujo de proceso de c-means difusos se enumera a continuación:
1.	Se supone un número fijo de conglomerados k.
2.	Inicialización : inicialice aleatoriamente las k medias $μ_k$ asociadas con los grupos y calcule la probabilidad de que cada punto de datos $x_i$ sea miembro de un grupo determinado k , P (el punto $x_i$ tiene la etiqueta k | $x_i$, k).
3.	Iteración :  calcular el centroide del clúster como el centroide ponderado dadas las probabilidades de pertenencia de todos los puntos de datos $x_i$:

4.	Terminación : Repetir hasta que se alcance el número máximo de iteraciones (dado por el usuario), o cuando el algoritmo haya convergido (es decir, que el cambio de los coeficientes entre dos iteraciones no sea superior a ϵ, el umbral de sensibilidad dado), la iteración puede quedar atrapada en algunos máximos o mínimos locales. Este algoritmo también minimiza la varianza intraclúster; como se dijo anteriormente el mínimo es un mínimo local, pot lo que los resultados dependen de la elección inicial; diferentes inicializaciones pueden conducir a resultados diferentes.  \

b) *Función $ \mu _2(x)=\dfrac{k(x-a)^{2}}{1+k(x-a)^{2}} $*

```python
def funcion2(x,k,a):
  num=len(x)
  m2= np.zeros(np.size(x))
  if k > 0:
    for i in range (1,num):
        if x[i]<= a:
            m2[i]=0;
        else:
            m2[i]=(k*(x[i]-a)**2)/(1+(k*(x[i]-a)**2));
  return m2
```

# Problema 3
*En la documentación del módulo **simpful** puedes encontrar la implementación del problema de la propina. Resuelve el problema 1 de la P06 con este módulo.*

```python
from simpful import *

# Se crea el sistema difuso
FS = FuzzySystem()

# Define los conjuntos difusos del sistema y las funciones de membresía de cada uno
S_0 = FuzzySet(function=Triangular_MF(a=0, b=0, c=2.5), term="pesimo")
S_1 = FuzzySet(function=Triangular_MF(a=0, b=2.5, c=5), term="malo")
S_2 = FuzzySet(function=Triangular_MF(a=2.5, b=5, c=7.5), term="regular")
S_3 = FuzzySet(function=Triangular_MF(a=5, b=7.5, c=10), term="bueno")
S_4 = FuzzySet(function=Triangular_MF(a=7.5, b=10, c=10), term="excelente")
FS.add_linguistic_variable("Servicio", LinguisticVariable([S_0,S_1, S_2, S_3,S_4], concept="Calidad del servicio", universe_of_discourse=[0,10]))

F_0 = FuzzySet(function=Triangular_MF(a=0, b=0, c=2.5), term="horrible")
F_1 = FuzzySet(function=Triangular_MF(a=0, b=2.5, c=5), term="mala")
F_2 = FuzzySet(function=Triangular_MF(a=2.5, b=5, c=7.5), term="regular")
F_3 = FuzzySet(function=Triangular_MF(a=5, b=7.5, c=10), term="buena")
F_4 = FuzzySet(function=Triangular_MF(a=7.5, b=10, c=10), term="deliciosa")
FS.add_linguistic_variable("Comida", LinguisticVariable([F_0,F_1, F_2,F_3,F_4], concept="Calidad de la comida", universe_of_discourse=[0,10]))

# Define output fuzzy sets and linguistic variable
T_1 = FuzzySet(function=Triangular_MF(a=0, b=0, c=5), term="poca")
T_2 = FuzzySet(function=Triangular_MF(a=0, b=5, c=10), term="media")
T_3 = FuzzySet(function=Trapezoidal_MF(a=5, b=10, c=15, d=15), term="generosa")
FS.add_linguistic_variable("Tip", LinguisticVariable([T_1, T_2, T_3], universe_of_discourse=[0,15]))

# Define las reglas
R1 = "IF (Servicio IS pesimo) OR (Comida IS horrible) THEN (Tip IS poca)"
R2 = "IF (Servicio IS bueno) THEN (Tip IS media)"
R3 = "IF (Servicio IS excelente) OR (Comida IS deliciosa) THEN (Tip IS generous)"
FS.add_rules([R1, R2, R3])

FS.plot_variable("Servicio")
FS.plot_variable("Comida")
FS.plot_variable("Tip")

FS.set_variable("Servicio", 4)
FS.set_variable("Comida", 8)

print(FS.Mamdani_inference(["Tip"]))
```

```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt

from skfuzzy import control as ctrl
Comida=ctrl.Antecedent(np.arange(0,100),'Comida')
Servicio=ctrl.Antecedent(np.arange(0,100),'Servicio')
Propina=ctrl.Consequent(np.arange(0,15),'Propina')

Comida.automf(5)
Servicio.automf(5)

Propina["Demasiado_Pequeña"] = fuzz.trimf(Propina.universe,[0,0,3])
Propina["Pequeña"] = fuzz.trimf(Propina.universe,[3,4,5])
Propina["Promedio"] = fuzz.trimf(Propina.universe,[6,6,11])
Propina["Generosa"] = fuzz.trimf(Propina.universe,[12,12,15])

Comida.view()
Servicio.view()
Propina.view()

rule1 = ctrl.Rule(Servicio['good'] & Comida['good'], Propina['Generosa'])
rule2 = ctrl.Rule(Servicio['poor'] & Comida['poor'], Propina['Demasiado_Pequeña'])
rule3 = ctrl.Rule(Servicio['poor'] & Comida['mediocre'], Propina['Pequeña'])
rule4 = ctrl.Rule(Servicio['mediocre'] & Comida['poor'], Propina['Pequeña'])
rule5 = ctrl.Rule(Servicio['good'] & Comida['average'], Propina['Promedio'])
rule6 = ctrl.Rule(Servicio['good'] & Comida['decent'], Propina['Generosa'])
rule7 = ctrl.Rule(Servicio['poor'] & Comida['average'], Propina['Pequeña'])
rule8 = ctrl.Rule(Servicio['average'] & Comida['average'], Propina['Promedio'])
rule9 = ctrl.Rule(Servicio['average'] & Comida['decent'], Propina['Promedio'])
rule10 = ctrl.Rule(Servicio['average'] & Comida['good'], Propina['Generosa'])
rule11 = ctrl.Rule(Servicio['poor'] & Comida['good'], Propina['Promedio'])
rule12 = ctrl.Rule(Servicio['decent'] & Comida['good'], Propina['Generosa'])
rule13 = ctrl.Rule(Servicio['mediocre'] & Comida['good'], Propina['Promedio'])
rule14 = ctrl.Rule(Servicio['mediocre'] & Comida['mediocre'], Propina['Demasiado_Pequeña'])
rule15 = ctrl.Rule(Servicio['decent'] & Comida['decent'], Propina['Generosa'])

tipping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)

tipping.input['Comida'] = 6.5
tipping.input['Servicio'] = 8.7

tipping.compute()

print(tipping.output['Propina'])
Propina.view(sim=tipping)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import skfuzzy as fuzz

colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']

# Define three cluster centers
centers = [[4, 2],
           [1, 7],
           [5, 6]]

# Define three cluster sigmas in x and y, respectively
sigmas = [[0.8, 0.3],
          [0.3, 0.5],
          [1.1, 0.7]]

# Generate test data
np.random.seed(42)  # Set seed for reproducibility
xpts = np.zeros(1)
ypts = np.zeros(1)
labels = np.zeros(1)
for i, ((xmu, ymu), (xsigma, ysigma)) in enumerate(zip(centers, sigmas)):
    xpts = np.hstack((xpts, np.random.standard_normal(200) * xsigma + xmu))
    ypts = np.hstack((ypts, np.random.standard_normal(200) * ysigma + ymu))
    labels = np.hstack((labels, np.ones(200) * i))

# Visualize the test data
fig0, ax0 = plt.subplots()
for label in range(3):
    ax0.plot(xpts[labels == label], ypts[labels == label], '.',
             color=colors[label])
ax0.set_title('Test data: 200 points x3 clusters.')
```

En la gráfica se pueden notar claramente 3 agrupamientos, cada uno con un color diferente. Sin embargo, esto no sucede siempre que se evalúan datos específicos, razón por la que se hace uso de los métodos de ***clustering*** que brinda python. 

Para realizar los agrupamientos se declaran las variables que la función 'cmeans' necesita, para este ejemplo estos son: 
*   Los datos a utilizar (alldata).
*   Número de agrupamientos (ncenters).
*   El valor del exponente para FCM (2).
*   La diferencia entre dos valores de pertenencia consecutivos (error=0.005).
*   El número máximo de iteraciones (maxiter=1000).
*   Coeficiente de partición difusa (init=None).

Luego, se procede a realizar los agrupamientos. Para este proceso se hace uso de un ciclo 'for', el cual será el encargado de separar los puntos dados en diferentes grupos, identificándolos con diferentes colores, y aumentando el número de grupos que se forman con cada iteración.

Después, otro ciclo for se encargará de marcar los centros de cada nuevo agrupamiento, marcándolo con un punto rojo como puede observarse en las impresiones. Este punto se obtiene tras realizar una operación de media y representa el grado de pertenencia de un punto a un cierto grupo. Este punto será re-calculado en cada iteración.

Así, al finalizar las iteraciones, se cuenta con 9 imágenes donde se observan de izquiera a derecha y de arriba hacia abajo de 2 y hasta 10 centros, cada uno pertenenciente a un agrupamiento.

```python
# Set up the loop and plot
fig1, axes1 = plt.subplots(3, 3, figsize=(8, 8))
alldata = np.vstack((xpts, ypts))
fpcs = []

for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)

    # Store fpc values for later
    fpcs.append(fpc)

    # Plot assigned clusters, for each data point in training set
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_membership == j],
                ypts[cluster_membership == j], '.', color=colors[j])

    # Mark the center of each fuzzy cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')

    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc))
    ax.axis('off')

fig1.tight_layout()
```

El coeficiente de partición difusa es una métrica que nos dice qué tan limpiamente se describen nuestros datos mediante un determinado modelo. Este se define entre valores de 0 a 1, donde 1 es el mejor. 
En el proceso que se realizó en pasos anteriores, se puede observar que la muestra estudiada tiene 3 agrupamientos, de los cuales se pueden obtener de 2 a 9 grupos si se itera repetidamente. 

La gráfica a continuación nos muestra los valores de fpc obtenidos para cada agrupamiento formado. Se puede observar que el valor de fpc se acerca al valor de 1 durante las primeras dos iteraciones y después de ello desciende poco a poco conforme los agrupamientos aumentan, siendo hasta el valor de 10 agrupaientos donde levemente se alza hacia el valor de 1 de nuevo. Como se puede ver, el número de agrupamientos que muestra la respuesta más limpia y por lo tanto el más ideal será el de 3, pues es quien alcanza el mayor valor de fpc, y aunque este resultado no sea novedad, para casos en los que la respuesta no es tan clara desde el inicio, este método ayuda a guiarse para obtener una respuesta más clara.

Cabe notar que las iteraciones se comienzan a partir de 2 grupos debido a que si se tomara en cuenta uno solo, el valor por defecto de fpc sería igual a 1 por englobar todos los puntos.

```python
fig2, ax2 = plt.subplots()
ax2.plot(np.r_[2:11], fpcs)
ax2.set_xlabel("Number of centers")
ax2.set_ylabel("Fuzzy partition coefficient")
```

Una vez obtenidos los agrupamientos ideales se puede proceder a ***clasificar nuevos datos***, añadiendo más elementos al modelo existente, proceso conocido como "predicción". 

Como ahora se conoce que el modelo tiene 3 agrupaciones, se puede reconstruir un modelo con 3 grupos para predecir a qué grupo pertenecerá un conjunto de valores nuevos. En este nuevo modelo se tienen valores aleatorios en el orden de los centros, por lo tanto los mismos pueden cambiar de posición conforme los nuevos datos sean añadidos.

```python
# Regenerate fuzzy model with 3 cluster centers - note that center ordering
# is random in this clustering algorithm, so the centers may change places
cntr, u_orig, _, _, _, _, _ = fuzz.cluster.cmeans(
    alldata, 3, 2, error=0.005, maxiter=1000)

# Show 3-cluster model
fig2, ax2 = plt.subplots()
ax2.set_title('Trained model')
for j in range(3):
    ax2.plot(alldata[0, u_orig.argmax(axis=0) == j],
             alldata[1, u_orig.argmax(axis=0) == j], 'o',
             label='series ' + str(j))
ax2.legend()
```

Finalmente, se generan datos muestreados uniformemente sobre este campo y se clasifican mediante cmeans_predict, incorporándolos al modelo preexistente.

Esto se logra primero generando nuevos datos distribuídos entre el rango [0,10] en x y y, luego se usa el modelo de 3 grupos para predecir a cuál de ellos pertenecen los nuevos puntos, después de ello son mostrados en una gráfica, ya separados por color de acuerdo a los centros ya conocidos.

Como se puede observar, crear un modelo a partir de datos obtenidos de una prueba puede ser de gran utilidad para hacer experimentos posteriores, simplificando la forma en que los datos son evaluados al basarse en el grado de pertenencia a los grupos en los que fue basado el modelo. 

```python
# Generate uniformly sampled data spread across the range [0, 10] in x and y
newdata = np.random.uniform(0, 1, (1100, 2)) * 10

# Predict new cluster membership with `cmeans_predict` as well as
# `cntr` from the 3-cluster model
u, u0, d, jm, p, fpc = fuzz.cluster.cmeans_predict(
    newdata.T, cntr, 2, error=0.005, maxiter=1000)

# Plot the classified uniform data. Note for visualization the maximum
# membership value has been taken at each point (i.e. these are hardened,
# not fuzzy results visualized) but the full fuzzy result is the output
# from cmeans_predict.
cluster_membership = np.argmax(u, axis=0)  # Hardening for visualization

fig3, ax3 = plt.subplots()
ax3.set_title('Random points classified according to known centers')
for j in range(3):
    ax3.plot(newdata[cluster_membership == j, 0],
             newdata[cluster_membership == j, 1], 'o',
             label='series ' + str(j))
ax3.legend()

plt.show()
```
